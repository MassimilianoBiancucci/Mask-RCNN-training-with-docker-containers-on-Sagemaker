```python
from imgaug import augmenters as iaa
from imgaug import parameters as iap
```

```python
params = [
    iap.Choice([-3, 3]),
    iap.Uniform(-3, 3),
    iap.Uniform([-1, 0], 1),
    iap.Uniform((-1, 0), 1)
]
iap.show_distributions_grid(params)
```

```python
params = [
    iap.Normal(0, 1),
    iap.Normal(5, 3),
    iap.Normal(iap.Choice([-3, 3, 10]), 1),
    iap.Normal(iap.Uniform(-3, 3), 1)
]
iap.show_distributions_grid(params)
```

```python
params = [
    iap.Laplace(0, 1),
    iap.Laplace(5, 3),
    iap.Laplace(iap.Choice([-3, 6]), 0.5),
    iap.Laplace(iap.Uniform(-3, 6), 1)
]
iap.show_distributions_grid(params)
```

```python
params = [
    iap.ChiSquare(1),
    iap.ChiSquare(5),
    iap.ChiSquare(iap.Choice([1, 5])),
    iap.RandomSign(iap.ChiSquare(3))
]
iap.show_distributions_grid(params)
```

```python
params = [
    iap.Beta(0.5, 0.5),
    iap.Beta(2.0, 2.0),
    iap.Beta(1.0, 0.5),
    iap.Beta(0.5, 1.0)
]
iap.show_distributions_grid(params)
```

```python
params = [
    iap.DiscreteUniform(0, 10),
    iap.DiscreteUniform(-10, 10),
    iap.DiscreteUniform([-10, -9, -8, -7], 10),
    iap.DiscreteUniform((-10, -7), 10)
]
iap.show_distributions_grid(params)
```

```python
params = [
    iap.Poisson(1),
    iap.Poisson(2.5),
    iap.Poisson((8, 12)),
    iap.RandomSign(iap.Poisson(2.5))
]
iap.show_distributions_grid(params)
```

```python
parameters = [
    iap.Choice([iap.Normal(0, 1), iap.Normal(2.5, 0.5)])
]
iap.show_distributions_grid(parameters)
```

```python
parameters = [
    iap.Absolute(iap.Normal(0.3, 0.5)),
    iap.Clip(iap.Absolute(iap.Normal(0.3, 0.5)), 0, 1),
    iap.Absolute(iap.Normal(0.4, 0.25)),
    iap.Clip(iap.Absolute(iap.Normal(0.4, 0.18)), 0, 1)
]
iap.show_distributions_grid(parameters)
```

```python
parameters = [
     iap.Clip(iap.Poisson((0, 50)), -50, 255),
     iap.Clip(iap.Absolute(iap.Normal(0.2, 0.2)), 0, 1),
]
iap.show_distributions_grid(parameters)
```
